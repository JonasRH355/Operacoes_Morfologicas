# Exemplo Morfologia

#**Trabalho de operações Morfológicas**
  Este é um tramalho que visa entender algumas operações morfológicas a partir da biblioteca OpenCV. Essas operações são: erosion, dilation, gradient, opening e closing.

####**Explicando melhor o que são essas operações**
  As operações são baseadas em imagens, usadas normalmente em forma binária. Precisando assim, duas entradas para o código funcionar, a *imagem* e o *Kernel*.
 ( Em algumas didáticas pode-se chamar como **elemento estrturante** Também )


```python

import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

Depois de importar as bibliotecas, devemos definir quais serão as imagens desejadas, além de sua altura e largura.

Próximo passo é definir o Kernel!

```python
img = cv2.imread('j.png',0)
img_opening = cv2.imread('j_ruido.png',0)
img_closing = cv2.imread('j_furos.png',0)
altura= img.shape
largura = img.shape
#Definindo a imagem e o seu tamanho

kernel = np.ones((100, 100),np.uint8)
#definindo Kernel
```

##Após isso é que a mágica acontece!
Após definir a Imagem e o Kernel, basta utilizar as operações que a biblioteca disponibiliza para chegar no respectivo objetivo

###EROSÃO
**"método de afinamento da imagem"**. Cada pixel preto tem valor 0, e o pixel branca tem valor 1. O kernel passará pela imagem, e o pixel na imagem original será erodido para 0 se um ou mais dos pixels sob o kernel forem 0.

###DILATAÇÂO
Já o método de deliatação é o oposto de erosão. Ao invés de o pixel ser erodido para 0, será transformando em 1, criando assim uma dilatação do formato na imagem.



```python
erosion = cv2.erode(img,kernel,iterations = 12)
# Processo que transforma pixels brancos em pretos. "Afina a Imagem"

dilation = cv2.dilate(img,kernel,iterations = 12)
# Processo que transforma pixels pretos em brancos. "Amplia a imagem"
```

###GRADIENT
Gradient morfológico é uma operação que requer uma operação de dilatação menos uma operação de erosão. Muito útil quando você precisa do contorno de uma forma da imagem.

###OPENING
A opening é uma transformação morfológica mais avançada. A opening é uma operação que requer uma operação de erosão dentro de uma operação de dilatação. **É MUITO utilizada para retirar ruido no mundo de imagens**

###CLOSING
O Closing é a operação inversa de Opening e requer uma operação de dilatação dentro de uma operação de erosão.
O qual se usa quando quiser preencher ruidos em um objeto da imagem.

```python
gradient = cv2.morphologyEx(img, cv2.MORPH_GRADIENT, kernel)
# Processo que mostra o contorno de um objeto da imagem

opening = cv2.morphologyEx(img_opening, cv2.MORPH_OPEN, kernel)
# Processo que Retira ruidos de imagens

closing = cv2.morphologyEx(img_closing, cv2.MORPH_CLOSE, kernel)
# Processo que retira ruidos de objetos
```

Apresentação das imagens após o uso de cada operação

```python
cv2_imshow(img)
cv2_imshow(img_opening)
cv2_imshow(img_closing)
#Como é a imagem normal

cv2_imshow(erosion)
cv2_imshow(dilation)
cv2_imshow(opening)
cv2_imshow(closing)
cv2_imshow(gradient)
# Após as operações
```
